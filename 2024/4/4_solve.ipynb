```python
import os
import numpy as np
import re
```

```python
file_path = os.path.join(os.getcwd(), 'input.txt')
with open(file_path, 'r') as f:
    lines = f.readlines()
```

```python
for i in range(len(lines)):
    lines[i] = re.sub(r'[^A-Z]', '', lines[i].strip())
text = ''.join(lines)
```

```python
def find_xmas(text):
    # Convert to 2D grid
    size = int(len(text) ** 0.5)  # Square grid
    grid = np.array(list(text)).reshape(size, size)

    count = 0
    # Directions to check: horizontal, vertical, diagonal
    directions = [(0, 1), (1, 0), (1, 1), (1, -1)]  # right, down, diagonal right, diagonal left

    for i in range(size):
        for j in range(size):
            # Check each direction from current position
            for di, dj in directions:
                # Try both forward and backward
                for multiplier in [1, -1]:  # 1 for forward, -1 for backward
                    word = ''
                    for k in range(4):  # XMAS is 4 letters
                        ni = i + k * di * multiplier
                        nj = j + k * dj * multiplier
                        # Check if position is valid
                        if 0 <= ni < size and 0 <= nj < size:
                            word += grid[ni][nj]
                    if word == 'XMAS':
                        count += 1

    return count
```

```python
find_xmas(text)
```

```python
def find_x_mas_pattern(text):
    # Convert to 2D grid
    size = int(len(text) ** 0.5)  # Square grid
    grid = np.array(list(text)).reshape(size, size)

    count = 0
    for i in range(size - 2):  # -2 because we need space for the pattern
        for j in range(1, size - 1):  # We need a middle position for A
            # Check both MAS patterns in X shape
            # Case 1: MAS forward in both diagonals
            if (grid[i][j - 1] == 'M' and grid[i + 1][j] == 'A' and grid[i + 2][j + 1] == 'S' and
                    grid[i][j + 1] == 'M' and grid[i + 2][j - 1] == 'S'):
                count += 1
            # Case 2: MAS backward (SAM) in both diagonals    
            elif (grid[i][j + 1] == 'S' and grid[i + 1][j] == 'A' and grid[i + 2][j - 1] == 'M' and
                  grid[i][j - 1] == 'S' and grid[i + 2][j + 1] == 'M'):
                count += 1
            # Case 3: Left diagonal forward (MAS), right diagonal backward (SAM)
            elif (grid[i][j - 1] == 'M' and grid[i + 1][j] == 'A' and grid[i + 2][j + 1] == 'S' and
                  grid[i][j + 1] == 'S' and grid[i + 2][j - 1] == 'M'):
                count += 1
            # Case 4: Left diagonal backward (SAM), right diagonal forward (MAS)
            elif (grid[i][j + 1] == 'M' and grid[i + 1][j] == 'A' and grid[i + 2][j - 1] == 'S' and
                  grid[i][j - 1] == 'S' and grid[i + 2][j + 1] == 'M'):
                count += 1

    return count
```

```python
find_x_mas_pattern(text)
```
