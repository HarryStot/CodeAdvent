```python
import os
import re
```

```python
file_path = os.path.join(os.getcwd(), 'input.txt')
with open(file_path, 'r') as f:
    lines = f.readlines()
```

```python
data = [line.strip() for line in lines]
```

```python
pattern = r'mul\((\d{1,3}),(\d{1,3})\)'
```

```python
results = []
for line in data:
    match = re.search(pattern, line)
    if match:
        print(f"Found match: {match.groups()}")
        a, b = map(int, match.groups())
        results.append(a * b)
```

```python
sum(results)
```

### Part 2

```python
pattern = r'(?P<action>do|don\'t)\(\)|mul\((\d{1,3}),(\d{1,3})\)'
enabled = True
results = []

for line in data:
    matches = re.finditer(pattern, line)
    for match in matches:
        if match.group('action'):
            enabled = match.group('action') == 'do'
        elif enabled and match.group(2):
            a, b = map(int, match.groups()[1:])
            results.append(a * b)
```

```python
sum(results)
```
