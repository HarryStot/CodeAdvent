```python
import os
import numpy as np
```

```python
file_path = os.path.join(os.getcwd(), 'input.txt')
with open(file_path, 'r') as f:
    lines = f.readlines()
```

```python
# Line format : '12345   12345\n'
data = []
for line in lines:
    values = line.strip().split()
    if len(values) == 2:  # Ensure we have a pair of numbers
        data.append([int(values[0]), int(values[1])])

data = np.array(data)
A = data[:, 0]
B = data[:, 1]
```

```python
diff = []
remaining_A = A.copy()
remaining_B = B.copy()
for i in range(len(A)):
    # Check if the array is empty
    if remaining_A.size == 0 or remaining_B.size == 0:
        break

    # Find the minimum values in each column
    min_a_idx = np.argmin(remaining_A)
    min_b_idx = np.argmin(remaining_B)

    # Calculate the absolute difference and append to results
    diff.append(abs(remaining_A[min_a_idx] - remaining_B[min_b_idx]))

    # Remove the rows with the minimum values to avoid finding them again
    remaining_A = np.delete(remaining_A, min_a_idx)
    remaining_B = np.delete(remaining_B, min_b_idx)

```

```python
print(sum(diff))
```

```python
similarity_score = 0
for a in A:
    if a in B:
        count = np.sum(B == a)
        similarity_score += count * a

print(similarity_score)
```
