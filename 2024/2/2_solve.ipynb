```python
import os
```

```python
file_path = os.path.join(os.getcwd(), 'input.txt')
with open(file_path, 'r') as f:
    lines = f.readlines()
```

```python
# Line format : '12 34 56 78\n'
data = []
for line in lines:
    values = line.strip().split()
    for i in range(len(values)):
        values[i] = int(values[i])
    data.append(values)
```

```python
safe_report = 0
def is_valid_sequence(seq):
    if len(seq) < 2:
        return False

    values_are_increasing = None
    last_value = seq[0]

    for value in seq[1:]:
        if value > last_value:
            if values_are_increasing is None:
                values_are_increasing = True
            elif values_are_increasing is False:
                return False
        elif value < last_value:
            if values_are_increasing is None:
                values_are_increasing = False
            elif values_are_increasing is True:
                return False

        diff = abs(value - last_value)
        if diff < 1 or diff > 3:
            return False

        last_value = value

    return values_are_increasing is not None

for line in data:
    if is_valid_sequence(line):
        safe_report += 1
        continue
```

```python
safe_report
```

### Part 2

```python
safe_report = 0
for line in data:
    # Check if original sequence is valid
    if is_valid_sequence(line):
        safe_report += 1
        continue

    # Try removing each value once
    for i in range(len(line)):
        test_sequence = line[:i] + line[i + 1:]
        if is_valid_sequence(test_sequence):
            safe_report += 1
            break
```
